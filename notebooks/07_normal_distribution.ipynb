# 標準正規分布

```python
import numpy as np
from scipy.stats import norm

%matplotlib inline
import matplotlib.pyplot as plt

# 平均が0，標準偏差が1の正規分布を特に標準正規分布という
x = np.arange(-5, 5, 0.01)
y = norm.pdf(x, 0, 1)
y /= np.sum(y)

plt.plot(x, y)
plt.show()
```

```python
# 標準偏差プラスマイナス1の範囲でだいたい70%のデータが含まれる
print('in S.D 1: {:.3}'.format(
    np.sum(
        y[tuple([np.where((-1 < x) & (x < 1))])])))

# 標準偏差プラスマイナス2の範囲でだいたい95%のデータが含まれる
print('in S.D 2: {:.3}'.format(
    np.sum(
        y[tuple([np.where((-2 < x) & (x < 2))])])))
```

# 一般の正規分布

```python
# 平均が0，標準偏差が1の正規分布を特に標準正規分布という
mean = 3
stddev = 5
range_ = 20

x = np.arange(-range_ + mean, range_ + mean, 0.01)
y = norm.pdf(x, loc=mean, scale=stddev)
y /= np.sum(y)

plt.plot(x, y)
plt.show()
```

```python
# 平均を基準として，標準偏差プラスマイナスS.D(1)の範囲でだいたい70%のデータが含まれる
print('in S.D stddev({}): {:.3}'.format(
    stddev,
    np.sum(y[tuple([np.where(((mean - stddev) < x) & (x < (mean + stddev)))])])))

# 平均を基準として，標準偏差プラスマイナスS.D(2)の範囲でだいたい95%のデータが含まれる
print('in S.D double stddev({}): {:.3}'.format(
    stddev * 2,
    np.sum(y[tuple([np.where(((mean - stddev * 2) < x) & (x < (mean + stddev * 2)))])])))
```

# 身長データとフィッティング（力技）

```python
tall_by_bin = {'143': 1, '148': 6, '153': 19, '158': 30, '163': 18, '168': 6}
tall_x = [float(e) for e in sorted(tall_by_bin.keys())]
tall_y = [float(tall_by_bin[b]) for b in sorted(tall_by_bin.keys())]

x = np.arange(143, 168, 0.01)
y = norm.pdf(x, loc=157.75, scale=5.4)
y *= 4.0e2

plt.plot(tall_x, tall_y)
plt.plot(x, y)
plt.show()
```
